```python
import pandas as pd
import os
import ast
```

```python
folder_path = r"..\csv"
print(os.listdir(folder_path))
```

```python
# Import de la table
games = pd.read_csv(r'..\csv\games.csv')
language_supports = pd.read_csv(r'..\csv\language_supports.csv')
language = pd.read_csv(r'..\csv\language.csv')
language_supports_type = pd.read_csv(r'..\csv\language_supports_type.csv')
game_genre = pd.read_csv(r'..\csv\game_genre.csv')
genre_id = pd.read_csv(r'..\csv\genre_id.csv')
game_game_type = pd.read_csv(r'..\csv\game_game_type.csv')
game_types = pd.read_csv(r'..\csv\game_types.csv')
players = pd.read_csv(r'..\csv\game_player_perspective_exploded.csv')
playeur_persp = pd.read_csv(r'..\csv\player_perspectives.csv')
```

```python
# --------  Games --------
# Supprimer les doublons
games = games.drop_duplicates()
# Drop de la colonne avec trop de missing
games.drop('total_rating', axis=1, inplace=True)
games
```

```python
# --------  game_genre --------
# Transformer les chaines en listes pour explode après
game_genre['genres'] = game_genre['genres'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else [])
game_genre = game_genre.explode('genres')
game_genre['genres'] = game_genre['genres'].astype(float)
# renommer id en game_id 
game_genre = game_genre.rename({'id': 'game_id', 'genres': 'genres_id'}, axis=1)
game_genre
```

```python
# -------------------------------
game_languages = pd.merge(language_supports, language, how='left', left_on='language', right_on='id')
game_langue = pd.merge(game_languages, language_supports_type, how='left', left_on='language_support_type', right_on='id')
# renommer colonne id_x et id_y
game_langue.rename(columns={'id' : 'language_type_id', 'id_x': 'language_support_id', 'id_y': 'language_id'}, inplace=True)
game_langue
```

```python
jeux = pd.merge(games, game_langue, how='left', right_on='game', left_on ='id')

jeux = jeux.drop(columns = ['language_support_id','language', 'language_id', 'language_support_type',
                            'native_name', 'locale', 'language_type_id','game'])
# renommer colonne
jeux = jeux.rename(columns = {'name_x':'langue', 'name_y': 'support_type'})
jeux
```

```python
# -
genres = pd.merge(game_genre, genre_id, how='left', left_on='genres_id', right_on='id')
genres.drop('id', axis = 1, inplace=True)
genres
```

```python
# --
jeu = pd.merge(jeux,genres, how = 'left', left_on = 'id', right_on = 'game_id')
jeu = jeu.drop('game_id', axis = 1)
```

```python
jeud = jeu.groupby('id').agg({'langue': lambda x: list(set(x)) if x.dropna().size > 0 else [], 
                                'support_type': lambda x: list(set(x)) if x.dropna().size > 0 else [], 
                                'name_x': 'first',
                                'slug':'first',
                                'summary': 'first',
                                'first_release_date':'first',
                                'genres_id': lambda x: list(set(x)) if x.dropna().size > 0 else [], 
                                'name_y': lambda x: list(set(x)) if x.dropna().size > 0 else []
                                }).reset_index()
```

```python
jeud = jeud.rename(columns={'name_x':'name', 'name_y':'genres'})
```

```python
game_type = pd.merge(game_game_type, game_types, how='left', left_on='game_type', right_on='id')
game_type.drop(columns=['game_type', 'id_y'], inplace=True)
```

```python
jeu_game = pd.merge(jeud, game_type, how='left', left_on='id', right_on='id_x')
jeu_game.drop(columns = ['id_x'], inplace=True)
```

```python
player = pd.merge(players, playeur_persp, how='left', left_on='player_perspectives', right_on='id')
player.drop(columns=['player_perspectives', 'id','slug'], inplace=True)
player = player.groupby('game_id').agg({'name': lambda x: list(set(x)) if x.dropna().size > 0 else []}).reset_index()
```

```python
game_all = pd.merge(jeu_game, player, how='left', left_on='id', right_on='game_id')
game_all.drop('game_id', axis = 1, inplace=True)
game_all = game_all.rename(columns={'name_x':'game_name', 'name_y':'players_perspectives'})
```

```python
# suppression de la ligne vide car pas de nom
game_all = game_all.dropna(subset=['game_name'])
game_all['type'] = game_all['type'].fillna('unknown')
game_all['type'] = game_all['type'].apply(lambda x: [x])
game_all['players_perspectives'] = game_all['players_perspectives'].apply(lambda x: x if isinstance(x, list) else [])
# cjnger format des colonnes de game_all
game_all['first_release_date'] = pd.to_datetime(game_all['first_release_date'], errors='coerce')
game_all['year'] = game_all['first_release_date'].dt.year
def year_to_period(year):
    if pd.isna(year):
        return 'Unknown'
    elif year < 2000:
        return 'Before 2000'
    elif year < 2010:
        return '2000s'
    elif year < 2020:
        return '2010s'
    else:
        return '2020s'

game_all['release_period'] = game_all['year'].apply(year_to_period)
 # Pour compatibilité avec MultiLabelBinarizer
game_all['release_period'] = game_all['release_period'].apply(lambda x: [x]) 
```

```python
games_all = game_all.copy()
```

```python
games_all.to_csv(r'..\csv\games_all.csv', index=False)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
import pandas as pd

# Liste des colonnes à inclure dans la recommandation
list_columns = ['genres', 'players_perspectives', 'release_period', 'type']
encoded_features = []

# Encodage binaire de chaque colonne
for col in list_columns:
    mlb = MultiLabelBinarizer()
    encoded = mlb.fit_transform(games_all[col])
    encoded_df = pd.DataFrame(encoded, columns=[f"{col}_{cls}" for cls in mlb.classes_])
    encoded_features.append(encoded_df)

# Fusion de tous les encodages
features_matrix = pd.concat(encoded_features, axis=1)
features_matrix.index = games_all['game_name']
```

```python
from sklearn.neighbors import NearestNeighbors

# Modèle KNN
knn_model = NearestNeighbors(metric='cosine', algorithm='auto')
knn_model.fit(features_matrix)
```

```python
def recommend_games(game_name, n=5):
    if game_name not in features_matrix.index:
        print(f"Le jeu '{game_name}' n'est pas dans le dataset.")
        return []
    
    game_vector = features_matrix.loc[game_name].values.reshape(1, -1)
    distances, indices = knn_model.kneighbors(game_vector, n_neighbors=n+1)
    
    # Indice 0 est le jeu lui-même → on l'exclut
    recommended_indices = indices.flatten()[1:]
    recommended_games = features_matrix.index[recommended_indices]
    
    print(f" Recommandations similaires à '{game_name}' :\n")
    for i, name in enumerate(recommended_games, 1):
        print(f"{i}. {name}")
    
    return recommended_games.tolist()
```

```python
recommend_games("Baldur's Gate", n=5)
```

```python
for metric in ['cosine', 'jaccard', 'hamming']:
    print(f"\n=== Suggestions avec {metric} ===")
    knn_model = NearestNeighbors(metric=metric, algorithm='brute')
    knn_model.fit(features_matrix)
    print(recommend_games("Thief II: The Metal Age", n=5))
```
